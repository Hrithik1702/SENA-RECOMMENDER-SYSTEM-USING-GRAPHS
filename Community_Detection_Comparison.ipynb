```python
import random
import networkx as nx
import numpy as np
from community import community_louvain
import matplotlib.pyplot as plt
```

```python
def no_of_nodes(path):
  m=0
  f = open(path,'r+')
  lines = f.readlines()
  for line in lines:
        t = line.split()
        u = t[0]
        i = t[1]
        m = max(m,int(u),int(i))
  return m
def generate_network(path):
    graph={}
    f = open(path,'r+')
    lines = f.readlines()
    for line in lines:
        t = line.split()
        u = t[0]
        i = t[1]
        graph[u] = graph.get(u,[])+[i]
        graph[i] = graph.get(i,[])+[u]
    G = nx.MultiGraph(graph)
    return G
```

```python
n = no_of_nodes('trust.txt')
print(n)
G = generate_network('trust.txt')
comms = community_louvain.best_partition(G)
no_of_communities=0
l={}
for u,v in comms.items():
  l[v]=1
  no_of_communities = max(no_of_communities,v)
print(no_of_communities)
print(max(l.keys()))
```

```python
def generate_network_2(path):
    graph = {}
    m=0
    f = open(path,'r+')
    lines = f.readlines()
    for line in lines:
        t = line.split()
        u = t[0]
        i = t[1]
        graph[u] = graph.get(u,[])+[i]
    H = nx.DiGraph(graph)
    G = ig.Graph.from_networkx(H)
    return G
```

```python
import leidenalg as la
G=generate_network_2('trust.txt')
partition = la.find_partition(G, la.ModularityVertexPartition)
partition.summary()
```
